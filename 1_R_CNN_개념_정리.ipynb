# <strong>2. R-CNN의 구조와 절차</strong>
##■ 구조

image.png

##■ 절차
1) 데이터와 레이블을 투입

    -. 데이터 : Image
    -. 레이블 : Bounding Box
   
2) 영역 제안 (Region Proposal)

    -. Bounding Box가 어디에 위치해야 하는지 신경망이 제안

3) 제안된 영역을 CNNs 에 입력

    -. Input값의 크기가 고정되어있음.
    -. 따라서 여러 크기의 Bounding Box(직사각형, 정사각형)의 크기를 맞추어 줌.
    
4) 분류 및 Bounding Box 조정

    ① 분류 (SVM)
    ② Bounding Box Regression
