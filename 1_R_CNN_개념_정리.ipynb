##■ 절차  
###1) 데이터와 레이블을 투입  
　-. 데이터 : Image  
　-. 레이블 : Bounding Box
  
　* Bounding Box 학습 과정  
　　① 선택적 탐색(Selective Search)를 통해 임의의 Bounding Box 설정  
　　② 임의의 Bounding Box와 정답 Bounding Box 의 IOU 계산  
　　　(IOU : Intersection Over Union → [영역의 교집합 / 영역의 합집합] )  
　　③ IOU가 특정 값 이상이 되도록 임의 영역을 조정 (IOU >= 0.5)  
   
###2) 영역 제안 (Region Proposal) - (선택적 탐색)  
　-. Bounding Box가 어디에 위치해야 하는지 신경망이 제안  
image.png


　■ R-CNN 에서의 선택적 탐색의 절차  
　　① 선택적 탐색은 초기의 작은 크기의 세분화 영역을 설정  
image.png

　　② 계층적 그룹 알고리즘을 사용하여 병합  
　　③ 이를 바탕으로 영역을 제안  

###3) 제안된 영역을 CNNs 에 입력  
　-. Input값의 크기가 고정되어있음.  
　-. 따라서 여러 크기의 Bounding Box(직사각형, 정사각형)의 크기를 맞추어 줌.  
    
###4) 분류 및 Bounding Box 조정

　② Bounding Box Regression
